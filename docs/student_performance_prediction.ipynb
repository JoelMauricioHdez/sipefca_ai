## **Planteamiento de la Problematica**

Para el **Sistema de Proyección de Desempeño Académico para Fortalecimiento de Sistemas de Consejería Académica (SIPEFCA)** se busca poder proyectar el desempeño de los estudiantes a lo largo de su vida académica en alguna institución académica. Para ello se deberán tomar varios factores en cuenta, puesto que tratar de hacer un proyección basada unicamente en las calificaciones logradas por el estudiante no proporcionará información valiosa para el estudiante o los consejeros.

En vista de esto, tenemos que tomar en cuenta más variables en cuenta. Para los fines del proyecto, el análisis se basará en las competencias presentes en cada asignatura. Esto se hará por medio del conteo del peso de cada competencia en las asignaturas. Este peso será contabilizado en base a la presencia de las subcompetencias de cada competencia en las distintas asignaciones del programa de las asignaturas. Y esto para asignatura dentro del pensum cursado por el estudiante.

Como base, el modelo tomará las competencias utilizadas por el ABET. Estas son las competencias utilizadas dentro de las carreras de ingenería del Instituto Tecnológico de Santo Domingo (INTEC). Estas serán las competencias que tomaremos como base para el entrenamiento del modelo para la generación de las proeycciones.

## **El modelo de predicción**

Una vez definido los factores a tomar en cuenta a la hora de realizar el análisis, ya podemos plantear cómo trabajará el modelo. En este jupyter notebook estaremos haciendo la prueba de concepto del modelo de análisis.

El modelo se basará en una mezcla de dos formas de análisis:

* **Pronóstico de series de tiempo (Time-Series Forecast)**

* **Regresión multivariable (Multivariate regression)**


### *Pronóstico de series de tiempo (Time-Series Forecasting) [2]*

Es el proceso del uso de modelos estadísticos para predecir valores futuros de una serie de tiempo (time-series) basada en resultados anteriores. Una serie de tiempo es una serie de datos recolectados en intervalos de tiempo.

Este tipo de análisis es aplicado en estadídica, finanzas y en empresas.

Sus componentes son:

* Moda (Trend). Muestra la dirección general de la serie en un periodo largo de tiempo.

* Estacionalidad (Seasonality). Muestra una mda que se repite en relación a una estasión particular.

* Componente Cíclico. Son modas que no poseen una tendencia de repetición sobre un periodo de tiempo. Pero si ocurren con el tiempo.

* Variación irregular. Son las fluctuaciones en la serie en la que no ocurren las modas o variaciones cíclicas conocidas.

* Descomposición ETS. Se usa para separar los componentes de una serie


### *Regresión Multivariable (Muiltivariate Regression) [3]*

Se trata de la técnica estadísitca para predecir el resultado de una variable basado en los valores de una o más variables. En esta contamos con la variable independiente, de la cuál queremos la predicción. Y las variables dependientes, de las cuales sale la predicción.

### 2. Pre procesamiento de la data

````python
# pre procesamiento de la data

def pre_process_student_data(student_data: str) -> pd.DataFrame:
    """
    Transform the nested JSON student data into a flat DataFrame suitable
    for analysis.
    """
    data = json.loads(student_data)

    rows = []
    student_id = data['student_id']

    for period in data['periods']:
        for course in period['courses']:
            # Convert letter grade to numerical value
            grade_map = {'A': 4.0, 'B+': 3.5, 'B': 3.0,
                        'C+': 2.5, 'C': 2.0, 'D+': 1.5, 'D': 1.0, 'F':-1.0,
                         'R': 0.5, None: 0}

            # only for test data that contains generated competences with gpt

            comp_map = {'C9':'C3', 'C8':'C2', 'C1':'C1', 'C10':'C4', 'C7':'C1',
                        'C5':'C5', 'C2':'C2', 'C6':'C6', 'C3':'C3', 'C4':'C4'}

            grade_value = grade_map[course['grade']]

            # Calculate weighted competency score
            for competency in course['competences']:

              row = {
                  'student_id': student_id,
                  'period': period['period'],
                  'course_id': course['course_id'],
                  'credits': course['credits'],
                  'numerical_grade': grade_value,
                  'competency': competency['comp_id'],
                  'weight': competency['weight'],
                  'period_course_load': period['cant_courses'],
                  'period_total_credits': period['total_credits']
              }
              rows.append(row)

    return pd.DataFrame(rows)

def get_relevant_data(student_data: str) -> Dict[str, any]:
  """
  Transform the nested JSON student data into a flat DataFrame suitable
  for analysis.
  """
  data =json.loads(student_data)

  result = {}

  rows = []
  student = data['student_id']

  for period in data['periods']:
        for course in period['courses']:
            # Convert letter grade to numerical value
            grade_map = {'A': 4.0, 'B+': 3.5, 'B': 3.0,
                        'C+': 2.5, 'C': 2.0, 'D+': 1.5, 'D': 1.0, 'F':-1.0,
                         'R': 0.5, None: 4}

            grade_value = grade_map[course['grade']]

            try:
              predicted_grade = grade_map[course['predicted_grade']]
            except:
              predicted_grade = 3.5

            # only for test data that contains generated competences via chat gpt

            comp_map = {'C9':'C3', 'C8':'C2', 'C1':'C1', 'C10':'C4', 'C7':'C1',
                        'C5':'C5', 'C2':'C2', 'C6':'C6', 'C3':'C3', 'C4':'C4'}

            # Calculate weighted competency score
            for comp in course['competences']:

              comp_id = ''
              if type(comp['comp_id']) != str:
                comp_id = comp['comp_id']['comp_id']
              else:
                comp_id = comp['comp_id']

              try:
                comp_performance = comp['performance']
              except:
                grade_percentage = grade_value / 4
                comp_performance = grade_percentage
                if comp_performance < 0.5:
                  comp_performance = 0.75

              comp_id = comp_map[comp_id]

              row = {
                  'year': period['period'].split('-')[0],
                  'period': period['period'].split('-')[1],
                  'course_id': course['course_id'],
                  'credits': course['credits'],
                  'numerical_grade': grade_value,
                  'predicted_grade': predicted_grade,
                  'competency': comp_id,
                  'weight': comp['weight'],
                  'comp_performance': comp_performance,
              }
              rows.append(row)

  return {"student_id":student, "data":pd.DataFrame(rows)}


other_data = get_relevant_data(pre_data)

processed_data = pre_process_student_data(pre_data)
````

````python
# visualizar calificaciones por periodo en el tiempo

def plot_student_grades_by_years(df: pd.DataFrame) -> pd.DataFrame:

  # Group by year and period, calculate the average grade
  avg_grades = df.groupby(["year", "period"]).agg({"numerical_grade": "mean"}).reset_index()

  # Create a Seaborn bar plot
  plt.figure(figsize=(10, 6))
  sns.barplot(data=avg_grades, x="year", y="numerical_grade", hue="period", palette="viridis")

  # Add labels and title
  plt.title("Average Grades Per Period Per Year", fontsize=16)
  plt.xlabel("Year", fontsize=12)
  plt.ylabel("Average Grade", fontsize=12)
  plt.legend(title="Period")
  plt.show()

  return avg_grades
````

##### 3.2.1 Proyecciones de desempeño de competencias.

Tomando en cuenta que queremos generar las proyecciones del desempeño del estudiante en las competencias para poder generar, a partís de esa data, las proyecciones del desempeño del estudiante en las asignaturas.

Para estos fines tenemos algunas formas de cómo realizar esto, pero para ello primero tenemos que revisar cómo estaremos utilizando la data que tenemos para lograrlo. Para poder generar las proyecciones podemos utilizar el desempeño de las siguientes maneras:

* tal y como está, utilizando el desempeño en las competencia tal y como viene.

* Calculando el promedio pesado de estos valores, basándonos en el peso de cada competencia en la asignatura.

* Calculando el promedio pesado de estos valores, basándonos tanto en el peso como en la cantidad de créditos que la asignatura.

A continuación estaremos realizando la comparación de estas tres formas para tomar la que mejor sirva para la generación de las proyecciones. Para esto se tomará en cuenta que los valores conseguidos en la aproximación muestren cambios cercanos a la realidad en cuando a lo "aleatorio" que puedan ser.


````python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from statsmodels.tsa.arima.model import ARIMA
from statsmodels.tsa.statespace.sarimax import SARIMAX


# Establecemos una competencia objetivo para los fines de la comparación

_comp = "C6"

# --------------------------------------------------------------------------------

# hacemos una copia de nuestro set de datos agrupando en base a los principales valores
ndf = df.groupby(['year','period','course_id','competency']).mean().reset_index()

# cambiamos para que el periodo contenga tango el año como el periofo ej: "202402"
ndf["period"] = ndf["year"].astype(str) + ndf["period"].astype(str)

# eliminamos la columna del año
ndf = ndf.drop(columns=["year"])

# Eliminar las columnas de numerical_grade, predicted_grade, ya que no
# necesitamos estos valores por el momento

ndf = ndf.drop(columns=["numerical_grade","predicted_grade"])

# --------------------------------------------------------------------------------

# Acá definimos los 3 grupos a comparar. En los 3 se está realizando un promedio de estos datos

# data con promedio simple

ndf1 = ndf[ndf["competency"] == _comp].groupby(['period','competency']).agg({"comp_performance":"mean"}).reset_index()

# data con promedio con peso de la competencia por los creditos de la asignatura

ndf2 = ndf[ndf['competency'] == _comp].groupby(['period','competency']).apply(lambda x: np.average(x['comp_performance'], weights=(x['weight']*x['credits']))).reset_index()

ndf2.rename(columns={0: 'comp_performance'}, inplace=True)

# data con promedio con peso de la competencia unicamente

ndf3 = ndf[ndf['competency'] == _comp].groupby(['period','competency']).apply(lambda x: np.average(x['comp_performance'], weights=x['weight'])).reset_index()

ndf3.rename(columns={0: 'comp_performance'}, inplace=True)

# --------------------------------------------------------------------------------

# definimos una función para poder visualizar de forma gráfica los resultados de la comparación
def plot_ndf(ndf, ndf2, ndf3, x: str, y: str, title="Yes",xlabel='period',ylabel='comp_performance', dpi=100, color='blue', color2='red'):
  plt.figure(figsize=(16,5), dpi=dpi)
  plt.plot(ndf[x],ndf[y],color=color, label='average')
  plt.plot(ndf2[x],ndf2[y],color=color2,label="weighted avg")
  plt.plot(ndf3[x],ndf3[y],color="yellow",label="weighted avg no credits")
  plt.legend(loc='best')
  plt.gca().set(title=title,xlabel=xlabel,ylabel=ylabel)
  plt.show()

# visualizamos nuestros resultados

plot_ndf(ndf1,ndf2,ndf3, x='period', y='comp_performance', title=f"Competency {_comp} performance in comparison with avg's visualization",)
````

###### 3.2.1.1Probando los diferentes modelos para la generación de las proyecciones del desempenó en las competencias

````python
# función para delimitar nuestros datos

def get_df_sect(df: pd.DataFrame, condition: bool):
  return df[condition]
````

6. Preparamos la data para ser analizada por la regresión multivariable.

````python


# class TrendSeriesForcaster():
#   def __init__(self, target: object):
#     self.target = target

def get_ARIMA(train : pd.DataFrame, test : pd.DataFrame, order = (2,3,2), alpha = 0.05, target='comp_performance') -> pd.DataFrame | None:
    '''
    Función para el calculo de las proyecciones de series de tiempo ARIMA

    Parámetros:

    train - Es el DataFrame que contiene la parte seleccionada para ser la data de entrenamiento del modelo.

    test - Es el DataFrame que contiene la parte seleccionada para ser la data de pruebas del modelo.

    target - Es el nombre de la columna analizada para la generación de las predicciones.

    El método retorna:

    Un DataFrame con las proyecciones de series de tiempo ARIMA.
    '''
    try:
      # acá separamos nuestra data a analizar
      y = train[target]#self.target

      # y la utilizamos para ingresarla en el modelo
      ARIMAmodel = ARIMA(y, order = order,)

      ARIMAmodel = ARIMAmodel.fit()

      # Acá generamos otro set de datos que contenga las predicciones realizadas con
      # el modelo ARIMA
      y_pred_b = ARIMAmodel.get_forecast(len(test.index))
      y_pred_df_b = y_pred_b.conf_int(alpha = alpha)
      y_pred_df_b["Predictions"] = ARIMAmodel.predict(start = y_pred_df_b.index[0], end = y_pred_df_b.index[-1])
      y_pred_df_b["period"] = test.period.values
      y_pred_df_b.index = test.index
      # y_pred_out_b = y_pred_df_b["Predictions"]

      ARIMA_out = y_pred_df_b

      return ARIMA_out
    except Exception as error:
      raise ValueError(f'could not get ARIMA prediction for {target}:', error)

def get_ARMA(train : pd.DataFrame, test : pd.DataFrame, order = (1,0,1), alpha = 0.05, target='comp_performance') -> pd.DataFrame | None:
  '''
  Función para el calculo de las proyecciones de series de tiempo ARMA

  Parámetros:

  train - Es el DataFrame que contiene la parte seleccionada para ser la data de entrenamiento del modelo.

  test - Es el DataFrame que contiene la parte seleccionada para ser la data de pruebas del modelo.

  target - Es el nombre de la columna analizada para la generación de las predicciones.

  El método retorna:

  Un DataFrame con las proyecciones de series de tiempo ARMA.
  '''
  try:
    # acá separamos nuestra data a analizar
    y = train[target]#self.target]
    # print('y: ')
    # print(y.head())

    # y la utilizamos para ingresarla en el modelo
    ARMAmodel = SARIMAX(y, order = order)
    ARMAmodel = ARMAmodel.fit()

    # Acá generamos otro set de datos que contenga las predicciones realizadas con
    # el modelo ARMA
    y_pred = ARMAmodel.get_forecast(len(test.index)) # type: ignore
    # print('y_pred:')
    # print(y_pred)

    y_pred_df = y_pred.conf_int(alpha = alpha)
    # print('y_pred_df')
    # print(y_pred_df.head())

    y_pred_df["Predictions"] = ARMAmodel.predict(start = y_pred_df.index[0], end = y_pred_df.index[-1]) # type: ignore
    y_pred_df["period"] = test.period.values
    y_pred_df.index = test.index

    # print('y_pred_dg predictions')
    # print(y_pred_df.head())

    # y_pred_out = y_pred_df["Predictions"]

    # print(y_pred_df)

    ARMA_out = y_pred_df
    # print('ARMA_out')
    # print(y_pred_out.head())

    return ARMA_out
  except Exception as error:
    raise ValueError(f'could not get ARMA prediction for {target}:', error)


def get_SARIMAX(train : pd.DataFrame, test : pd.DataFrame, order = (5,4,2), alpha = 0.05, seasonal_order=(3,1,1,12),target='comp_performance') -> pd.DataFrame | None:
  '''
  Función para el calculo de las proyecciones de series de tiempo SARIMAX

  Parámetros:

  train - Es el DataFrame que contiene la parte seleccionada para ser la data de entrenamiento del modelo.

  test - Es el DataFrame que contiene la parte seleccionada para ser la data de pruebas del modelo.

  target - Es el nombre de la columna analizada para la generación de las predicciones.

  El método retorna:

  Un DataFrame con las proyecciones de series de tiempo SARIMAX.
  '''
  try:
    # acá separamos nuestra data a analizar
    y = train[target]#self.target]


    SARIMAXmodel = SARIMAX(y, order = order, seasonal_order=seasonal_order, trend='ct')
    SARIMAXmodel = SARIMAXmodel.fit()

    # Acá generamos otro set de datos que contenga las predicciones realizadas con
    # el modelo SARIMAX
    y_pred_c = SARIMAXmodel.get_forecast(len(test.index)) # type: ignore
    y_pred_df_c = y_pred_c.conf_int(alpha = alpha)
    y_pred_df_c["Predictions"] = SARIMAXmodel.predict(start = y_pred_df_c.index[0], end = y_pred_df_c.index[-1]) # type: ignore
    y_pred_df_c["period"] = test.period.values
    y_pred_df_c.index = test.index
    # y_pred_out_c = y_pred_df_c["Predictions"]

    # print(y_pred_df_c)
    SARIMAX_out = y_pred_df_c

    return SARIMAX_out
  except Exception as error:
    raise ValueError(f'could not get ARMA prediction for {target}:', error)


def get_rmse(test : pd.DataFrame, predictions : pd.DataFrame, target : str = 'comp_performance' , pred_target : str = "Predictions"):
  try:
    return np.sqrt(mean_squared_error(test[target].values, predictions[pred_target])) # type: ignore
  except Exception as error:
    raise ValueError('could not get the rmse for the current test data, ', error)

  # def __getitem__(self):
  #   return
````

````python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from statsmodels.tsa.arima.model import ARIMA
from statsmodels.tsa.statespace.sarimax import SARIMAX


class CompPerformancePredictor():
  '''
  Clase para la generación de las predicciones de desempeño del estudiante
  en las competencias.

  Parámetros:

  data - Es el DataFrame inicial con los datos historicos del estudiante.

  pred_data - Version formateada del set de datos para la generación de
  las proyecciones.
  '''
  def __init__(self, data: pd.DataFrame):
    self.data : pd.DataFrame = data
    self.pred_data : pd.DataFrame | None = self.get_weighted_avg(self.data)


  def forecast_competency_performance(self, df: pd.DataFrame | None = None):
    """
    Método para realizar las proyecciones de desempeño de las competencias
    basado en el Time Series Forecasting.

    Parámetros:
    df - Es un DataFrame opcional para pasar un set de datos con los datos
    de las predicciones de desempeño de las competencias.

    El método retorna:

    Un DataFrame con las proyecciones de desempeño de las competencias.

    Estructura del DataFrame:
    {
      index: int,
      period: str,
      competency: str,
      comp_performance: float
    }
    """
    if df is None:
      if self.pred_data is None:
        df = self.data
        df = self.get_weighted_avg(df)
      else:
        df = self.pred_data
    else:
      df = self.data
      df = self.get_weighted_avg(df)

    results = []

    for competency in df['competency'].unique(): # type: ignore
      # print('current_comp: ',competency)
      competency_regs = df[df['competency'] == competency].reset_index(drop=True) # type: ignore
      # print(f'comp {competency} regs: ')
      # print(competency_regs.head())

      # get predicted performance for the target period
      results.append(self.get_predicted_comp_performance(competency_regs))

    # print(results)

    merged_df = pd.concat(results)

    merged_df = merged_df.sort_values(by=['period']).reset_index(drop=True)

    # Return forecasted data as a DataFrame
    return merged_df

  def get_weighted_avg(self, df: pd.DataFrame | None = None):
    '''
    Retorna el set de datos agrupado por periodos y competencias,
    incluyendo el promedio "pesado" de cada competencia de las
    asignaturas existentes en un determinado perido.

    función necesaria para preparar los datos para la generación de las
    predicciones del desempeño académico del estudiante
    '''
    if df is None:
      df = self.data

    try:
      df = df[df['credits'] > 0]

      # print(df.groupby(['period','competency']))
      df = (df
            .groupby(['period','competency'])
            .apply(lambda x: np.average(x['comp_performance'], weights=(x['weight']*x['credits']))) # type: ignore
            .reset_index())

      df.rename(columns={0: 'comp_performance'}, inplace=True)

      return df
    except Exception as error:
      print(error)
      return None

  def get_predicted_comp_performance(self, df: pd.DataFrame, target: str = 'comp_performance', diff:int=2):
    '''
    Retorna el desempeño predecido del estudiante en de un determinado set de datos.

    El parametro "df" debe ser un DataFrame de pandas que contenga unicamente los
    valores de sobre la competencia de la cual se desean hacer las predicciones.
    '''
    if df is None:
      raise ValueError('df cannot be None')

    # print('Received df: ')
    # print(df)

    # define test and train data
    train = df[df.index <= len(df.period.values) - diff]
    # print('training data set: ')
    # print(train)

    test = df[df.index >= len(df.period.values) - diff]
    # print('testing data set: ')
    # print(test)

    # try to get predicted performance with each method type
    try:
      pred = self.get_prediction_rmse(train,test,target).copy() # type: ignore

      # print('pre test')
      # print(test.head())

      # print('pred')
      # print(pred.head)

      # # we get the predicted data and asign its values to the
      # for period in test['period'].unique():
      test.loc[test.period.isin(pred.period),'comp_performance'] = pred['Predictions'].values

      # print('test ')
      # print(test.head())
      # print()
      # print()
      # print()

      # test.drop(test.index == len(df.period.values) - diff)
      result = pd.concat([train,test[test.index != len(df.period.values) - diff]])
      # print('result')
      # print(result)
      # print()
      # print()
      # print()

      return result

    except ValueError as error:
      print("Error predicting the performance for the target competency,", error)
      return


  def get_prediction_rmse(self,train : pd.DataFrame, test: pd.DataFrame, target: str = 'comp_performance',):
    '''
    Método para las predicciones de los distintos modelos de series de tiempo
    y sus respectivos valores de rmse para comprobar el nivel de "certeza" de
    las predicciones.

    Parámetros:

    train - Es el DataFrame que contiene la parte seleccionada para ser la data de entrenamiento del modelo.

    test - Es el DataFrame que contiene la parte seleccionada para ser la data de pruebas del modelo.

    target - Es el nombre de la columna analizada para la generación de las predicciones.

    Este método retorna un un diccionario con las siguientes propiedades:

    - pred - El DataFrame con la información de las proyecciones generadas.
    - rmse - El valor del rmse de las predicciones.

    '''
    pred_a = {"pred":None,"rmse":1}
    pred_b = {"pred":None,"rmse":1}
    pred_c = {"pred":None,"rmse":1}

    try:
      # sf = TrendSeriesForcaster(target)

      pred_a['df'] = get_ARMA(train,test,target=target)
      # print("a: ",pred_a['df'].head())
      pred_a['rmse'] = get_rmse(test,pred_a['df'])
      # print('ARMA rmse: ',pred_a['rmse'])


      if pred_a['rmse'] > 0.05:
         lowest = pred_a

         pred_b['df'] = get_ARIMA(train,test,target=target)
        #  print("b: ",pred_b['df'].head())
         pred_b['rmse'] = get_rmse(test,pred_b['df']) # type: ignore
        #  print('ARIMA rmse: ',pred_b['rmse'])



         if pred_b['rmse'] > 0.05:
            if pred_b['rmse'] < lowest['rmse']:
              lowest = pred_b

            pred_c['df'] = get_SARIMAX(train,test,target=target)
            # print("c",pred_c['df'].head())
            pred_c['rmse'] = get_rmse(test,pred_c['df']) # type: ignore
            # print('SARIMAX rmse: ',pred_c['rmse'])



            if pred_c['rmse'] > 0.05:
              if pred_c['rmse'] < lowest['rmse']:
                lowest = pred_c
            else:
              return pred_c['df']
         else:
            return pred_b['df']
      else:
        return pred_a['df']
      return lowest['df']
    except Exception as error:
      raise ValueError('there was an error trying to get the predictions, ', error)
      return {"error": error}


````

````python
class CoursePerformancePredictor():
  '''
  Clase para el calculo de las proyecciones de desempeño académico de los estudiantes.

  Parámetros:

  data - Es el DataFrame inicial con los datos historicos del estudiante.

  predicted_comp_performance - Es el DataFrame con los datos de las proyecciones
  de desempeño de las competencias. Origindario de la clase CompPerformancePredictor.

  Métodos:

  forecast_courses_performance - Método para la generación de las proyecciones de
  desempeño académico en las distintas asignaturas de un periodo en particular.

  predict_course_performance - Método para la generación de las preddicciones de
  desempeño académico de una asignatura.

  '''

  def __init__(self, data: pd.DataFrame, predicted_comp_performance: pd.DataFrame):
    self.data = data
    self.predicted_comp_performance = predicted_comp_performance

  def forecast_courses_performance(self,period:str,df: pd.DataFrame | None = None):
    '''
    Método para la generación de las proyecciones de desempeño académico en las
    distintas asignaturas de un periodo en particular.

    Parámetros:
    period - Periodo del cual se desean generar las proyecciones.

    df - Es un DataFrame opcional para pasar un set de datos con los datos
    de las predicciones de desempeño de las competencias y el desempeño
    logrado por el estudiante en los periodos anteriores. El valor por
    defecto es el DataFrame original que posee el objeto
    CoursePerformancePredictor.

    Este método retorna un DataFrame que contiene los datos actualizados de
    las asignaturas del periodo seleccionado incluyendo el desempeño proyectado
    para cada una.
    '''
    if df is None:
      df = self.data

    period_df = df[df['period'] == period]

    courses_performance = []

    # print('courses ')
    # print(period_df['course_id'].unique())

    for course in period_df['course_id'].unique():
      courses_performance.append(self.predict_course_performance(course,df))

    return pd.concat(courses_performance)

  def predict_course_performance(self, course: str, df: pd.DataFrame | None = None,):
    '''
    Método para la generación de las preddicciones de desempeño académico de una
    asignatura.

    Parámetros:
    course - Cadena de texto que represente el código de asignatura.

    df - Es un DataFrame opcional para pasar un set de datos con los datos
    de las predicciones de desempeño de las competencias y el desempeño
    logrado por el estudiante en los periodos anteriores. El valor por
    defecto es el DataFrame original que posee el objeto
    CoursePerformancePredictor.

    Este método retorna un DataFrame con los datos de la asignatura en cuestión.
    '''
    if df is None:
      df = self.data

    course_data = df[df['course_id'] == course]
    period = course_data['period'].unique()[0]
    # print('period')
    # print(period)

    competences = []

    for competence in course_data['competency'].unique():
      weight = (course_data[course_data['competency'] == competence]['weight']
                .unique()[0])
      grade = (course_data[course_data['competency'] == competence]['numerical_grade']
               .unique()[0])

      pred_perf = self.predicted_comp_performance[self.predicted_comp_performance['competency'] == competence]
      pred_perf['weight'] = weight
      pred_perf['numerical_grade'] = grade

      competences.append(pred_perf[pred_perf['period'] == period])

    comp_df = pd.concat(competences)
    # print(comp_df)

    wavg_pred = self.__get_weighted_avg(comp_df) * 4
    # print("Weighted avg prediciton: ", self.get_weighted_avg(comp_df) * 4)
    mvlr_pred = self.__get_regressed_course_performance(comp_df, self.data)
    # print("Grade predicition: ", self.__get_regressed_course_performance(comp_df, self.data))

    course_data['predicted_grade'] = (wavg_pred + mvlr_pred) / 2
    course_data['predicted_grade'] = course_data['predicted_grade'].clip(upper=4.0)
    # course_data.clip(lower{'predicted_grade': 4.0}, inplace=True)

    return course_data

  def __get_weighted_avg(self, df: pd.DataFrame):
    '''
    Método para el cálculo del promedio pesado de las proyecciones de estudiantes.

    Esta permite calcular el desempeño académico directamente en base a las
    predicciones de desempeño en las competencias de la asignatura.

    Parámetros:
    df - Es un DataFrame que contiene los datos referentes a una asignatura
    determinada.

    Valor retornado:
    La función retorna un valor numérico que representa el desempeño académico
    proyectado de la asignatura en cuestión.
    '''
    comp_weight = 0
    weights = 0
    for competency in df['competency'].unique():
      comp_perf = df[df['competency'] == competency]['comp_performance'].unique()[0]
      weight = df[df['competency'] == competency]['weight'].unique()[0]
      comp_weight += comp_perf * weight
      weights += weight

    return round(comp_weight/weights, 2)

  def __get_regressed_course_performance(self, target: pd.DataFrame,df: pd.DataFrame | None = None):
    '''
    Esta función es para conseguir las predicciones de desempeño académico
    en la asignatura basado en las predicciones generadas de las competencias
    manejadas para el periodo.

    Parámetros:
    Target - Es un DataFrame que contiene los datos referentes a una asignatura
    determinada.

    df - Es un DataFrame opcional para pasar un set de datos con los datos
    de las predicciones de desempeño de las competencias y el desempeño
    logrado por el estudiante en los periodos anteriores.

    Valor retornado:
    La función retorna un valor numérico que representa el desempeño académico
    proyectado de la asignatura en cuestión.
    '''
    if df is None:
      df = self.data

    # comp_map = {'C1':1,'C2':2,'C3':3,'C4':4,'C5':5,'C6':6}

    # print('df')
    # print(df.head())

    # print('target')
    # print(target.head())

    # df['competency_'] = df['competency'].apply(lambda x: comp_map[x])

    X = df[['comp_performance','weight']]
    y = df['numerical_grade']

    regr = LinearRegression()
    regr.fit(X,y)

    target_X = []
    for competency in target['competency'].unique():
      t = target[target['competency'] == competency].copy()
      # print(t.head())
      target_weight = t['weight'].unique()[0]
      target_comp_perf = t['comp_performance'].unique()[0]

      target_X.append([target_comp_perf,target_weight])


    return round(regr.predict(target_X)[0],2)


````

````python
class CompetencyPerformance():
  '''
  Class for the generation of the competency related DataFrames with the data
  necesary to generate the json to be passed with the API for the frontEnd of
  SIPEFCA.

  To create an object from this class it is required to have a default
  DataFrame. This DataFrame will be used if another one is not passed
  as an argument in any of the methods.

  plotting is False by default. If you are testing the results from the
  dataFrames you simply need to set this variable to True.\n

  Because we would want to be able to visualize the data with both real
  and predicted data we have the option to pass a boolean to set if we
  want to add the predicted values to the processing.

  '''
  def __init__(self, df: pd.DataFrame, plotting: bool = False, use_predicted_values: bool = False):
    self.df = df
    self.plotting = plotting
    self.use_predicted_values = use_predicted_values

  def get_competences_by_period(self, period: str, df: pd.DataFrame | None = None) -> pd.DataFrame:
    '''
    Get the DataFrame with the info of the competences by the periods.

    This return a DataFrame you can either pass it as an argument for
    another method or simply convert it to JSON to sent as a API.
    '''
    if df is None:
      df = self.df

    _year = period[:4]
    _period = period[4:]

    comp_period_performance = df.groupby(["year","period"]).get_group((_year,_period))


    if self.plotting:

      if self.use_predicted_values:

        # Prepare data for stacked plot
        comparison_df = pd.melt(
            comp_period_performance,
            id_vars=["competency"],
            value_vars=["comp_performance", "predicted_grade"], # fix to predicted performance
            var_name="type",
            value_name="Performance"
        )


        # Create a stacked bar plot
        plt.figure(figsize=(12, 6))
        bar_plot = sns.barplot(
            data=comparison_df,
            x="competency",
            y="Performance",
            hue="type",
            palette="Set2"
        )


        # Add labels and title
        handles, labels = bar_plot.get_legend_handles_labels()

        plt.title(f"Comparison Between Real and Predicted Competency performance for {period}", fontsize=16)
        plt.xlabel("Competences", fontsize=12)
        plt.ylabel("Performance", fontsize=12)
        plt.legend(handles=handles, labels=["Real Performance", "Predicted performance"], loc='upper right')
        plt.show()


      else:

        # Create a Seaborn bar plot
        plt.figure(figsize=(10, 6))
        sns.barplot(data=comp_period_performance, x="competency", y="comp_performance", hue="competency", palette="viridis")

        # Add labels and title
        plt.title(f"Competency performance for period {period}", fontsize=16)
        plt.xlabel("Competeny", fontsize=12)
        plt.ylabel("Performance", fontsize=12)
        plt.show()

    return comp_period_performance

  def get_competences_by_year(self, year: str, df: pd.DataFrame | None = None) -> pd.DataFrame:
    '''
    Get the DataFrame with the info of the competences performance of a specific year.

    This return a DataFrame you can either pass it as an argument for
    another method or simply convert it to JSON to sent as a API.
    '''
    if df is None:
      df = self.df

       # Group by period and competency to calculate average performance
    competency_performance = df.groupby(["year","period", "competency"]).agg(
        {"comp_performance": "mean","predicted_grade":"mean"}
    ).reset_index()

    competency_performance = competency_performance[competency_performance["year"] == year]

    if (self.plotting):

      if self.use_predicted_values:

        # Prepare data for stacked plot
        comparison_df = pd.melt(
            competency_performance,
            id_vars=["competency"],
            value_vars=["comp_performance", "predicted_grade"], # fix to predicted performance
            var_name="type",
            value_name="Performance"
        )


        # Create a stacked bar plot
        plt.figure(figsize=(12, 6))
        bar_plot = sns.barplot(
            data=comparison_df,
            x="competency",
            y="Performance",
            hue="type",
            palette="Set2"
        )


        # Add labels and title
        handles, labels = bar_plot.get_legend_handles_labels()

        plt.title(f"Comparison Between Real and Predicted Competency performance for {year}", fontsize=16)
        plt.xlabel("Competences", fontsize=12)
        plt.ylabel("Performance", fontsize=12)
        plt.legend(handles=handles, labels=["Real Performance", "Predicted performance"], loc='upper right')
        plt.show()


      else:

        # Create a Seaborn bar plot
        plt.figure(figsize=(12, 6))
        sns.barplot(
            data=competency_performance,
            x="period",
            y="comp_performance",
            hue="competency",
            palette="Set2"
        )

        # Add labels and title
        plt.title(f"Performance on Competences by Period on {year}", fontsize=16)
        plt.xlabel("Period", fontsize=12)
        plt.ylabel("Average Competency Performance (%)", fontsize=12)
        plt.legend(title="Competency ID", bbox_to_anchor=(1.05, 1), loc='upper left')
        plt.tight_layout()
        plt.show()

    return competency_performance


  def get_global_competences(self, dfs) -> pd.DataFrame:
    '''
    Get the DataFrame with the info of tyhe global competence performance of the student

    IMPORTANT: To get the global competence performance you need to generate the DataFrames
    for each of the years. This way you pass an array with these DataFrames

    This return a DataFrame you can either pass it as an argument for
    another method or simply convert it to JSON to sent as a API.
    '''
    try:
      assert(type(dfs[0]) != pd.DataFrame)
    except Exception as e:
      print("dfs must contain an Array of pd.DataFrame")

    all_years_data = pd.concat(dfs, ignore_index=True)

    global_performance = (
        all_years_data.groupby("competency")["comp_performance"]
        .mean()
        .reset_index()
        .rename(columns={"comp_performance": "global_performance"})
    )

    if (self.plotting):

      if self.use_predicted_values:

        global_performance = (
            all_years_data.groupby("competency")
              .agg({"comp_performance":"mean","predicted_grade":"mean"})
              .reset_index()
              .rename(columns={
                  "comp_performance":"global_performance",
                  "predicted_grade":"global_predicted_performance"
                  })
        )
        # Prepare data for stacked plot
        comparison_df = pd.melt(
            global_performance,
            id_vars=["competency"],
            value_vars=["global_performance", "global_predicted_performance"], # fix to predicted performance
            var_name="type",
            value_name="Performance"
        )


        # Create a stacked bar plot
        plt.figure(figsize=(12, 6))
        bar_plot = sns.barplot(
            data=comparison_df,
            x="competency",
            y="Performance",
            hue="type",
            palette="Set2"
        )


        # Add labels and title
        handles, labels = bar_plot.get_legend_handles_labels()

        plt.title(f"Comparison Between Real and Predicted Global Competency performance", fontsize=16)
        plt.xlabel("Competences", fontsize=12)
        plt.ylabel("Performance", fontsize=12)
        plt.legend(handles=handles, labels=["Real Performance", "Predicted performance"], loc='upper right')
        plt.show()


      else:

        # Visualization: Global Performance for Each Competency
        plt.figure(figsize=(10, 6))

        sns.barplot(
            data=global_performance,
            x="competency",
            y="global_performance",
            hue='competency',
            palette="coolwarm"
        )

        # Adding titles and labels
        plt.title("Global Performance per Competency", fontsize=16)
        plt.xlabel("Competency ID", fontsize=12)
        plt.ylabel("Global Performance (%)", fontsize=12)

        # Display the plot
        plt.tight_layout()
        plt.show()

    return global_performance

  def get_competences_by_course(self, course: str, df: pd.DataFrame | None = None) -> pd.DataFrame:
    '''
    Get the DataFrame with the info of the competences performance on a specific course.

    This return a DataFrame you can either pass it as an argument for
    another method or simply convert it to JSON to sent as a API.
    '''
    if df is None:
      df = self.df

    course_competences = df.groupby(["course_id"]).get_group((course,))

    if (self.plotting):

      if self.use_predicted_values:

        # Prepare data for stacked plot
        comparison_df = pd.melt(
            course_competences,
            id_vars=["competency"],
            value_vars=["comp_performance", "predicted_grade"], # fix to predicted performance
            var_name="type",
            value_name="Performance"
        )

        # Create a stacked bar plot
        plt.figure(figsize=(12, 6))
        bar_plot = sns.barplot(
            data=comparison_df,
            x="competency",
            y="Performance",
            hue="type",
            palette="Set2"
        )

        # Add labels and title
        handles, labels = bar_plot.get_legend_handles_labels()

        plt.title(f"Comparison Between Real and Predicted Competency performance for {course}", fontsize=16)
        plt.xlabel("Competences", fontsize=12)
        plt.ylabel("Performance", fontsize=12)
        plt.legend(handles=handles, labels=["Real Performance", "Predicted performance"], loc='upper right')
        plt.show()


      else:


        # Create a Seaborn bar plot
        plt.figure(figsize=(10, 6))
        sns.barplot(data=course_competences, x="competency", y="comp_performance", hue="competency", palette="viridis")

        # Add labels and title
        plt.title(f"Competency performance for {course}", fontsize=16)
        plt.xlabel("Competences", fontsize=12)
        plt.ylabel("Performance", fontsize=12)
        plt.show()

    return course_competences
````

````python
class AcademicPerformance():
  '''
  Class for the generation of the academic related DataFrames with the data
  necesary to generate the json to be passed with the API for the frontEnd of
  SIPEFCA.

  To create an object from this class it is required to have a default
  DataFrame. This DataFrame will be used if another one is not passed
  as an argument in any of the methods.

  plotting is False by default. If you are testing the results from the
  dataFrames you simply need to set this variable to True.

  Because we would want to be able to visualize the data with both real
  and predicted data we have the option to pass a boolean to set if we
  want to add the predicted values to the processing.

  '''
  def __init__(self, df: pd.DataFrame, plotting: bool = False, use_predicted_values: bool = False):
    self.df = df
    self.plotting = plotting
    self.use_predicted_values = use_predicted_values

  def get_grades_by_year(self, year: str, df: pd.DataFrame | None = None) -> pd.DataFrame:
    '''
    Get the DataFrame with the info of the grade performance of a specific year.

    This return a DataFrame you can either pass it as an argument for
    another method or simply convert it to JSON to sent as a API.
    '''
    if df is None:
      df = self.df

    avg_grades = df.groupby(["year"]).get_group((year,))

    if self.plotting:

      if self.use_predicted_values:

        comparison_df = pd.melt(
            avg_grades,
            id_vars=["period"],
            value_vars=["numerical_grade", "predicted_grade"],
            var_name="type",
            value_name="Grade"
        )


        # Create a stacked bar plot
        plt.figure(figsize=(12, 6))
        bar_plot = sns.barplot(
            data=comparison_df,
            x="period",
            y="Grade",
            hue="type",
            palette="Set2"
        )


        # Add labels and title
        handles, labels = bar_plot.get_legend_handles_labels()

        plt.title("Comparison Between Real and Predicted Grade Percentages", fontsize=16)
        plt.xlabel("Periods", fontsize=12)
        plt.ylabel("Grade", fontsize=12)
        plt.legend(handles=handles, labels=["Real Grade", "Predicted Grade"], loc='upper right')
        plt.show()

      else:

        # Create a Seaborn bar plot
        plt.figure(figsize=(10, 6))
        sns.barplot(data=avg_grades, x="period", y="numerical_grade", hue='period', palette="viridis")

        # Add labels and title
        plt.title(f"Grades on {year}", fontsize=16)
        plt.xlabel("Period", fontsize=12)
        plt.ylabel("Average Grade", fontsize=12)
        plt.show()

    return avg_grades

  def get_grades_by_period(self, period: str, df: pd.DataFrame | None = None) -> pd.DataFrame:
    '''
    Get the DataFrame with the info of the grade performance of a specific period.

    This return a DataFrame you can either pass it as an argument for
    another method or simply convert it to JSON to sent as a API.
    '''
    if df is None:
      df = self.df

    _year = period[:4]
    _period = period[4:]

    period_grades = df.groupby(["year","period"]).get_group((_year,_period))

    if self.plotting:

      if self.use_predicted_values:
        # Prepare data for stacked plot
        comparison_df = pd.melt(
            period_grades,
            id_vars=["course_id"],
            value_vars=["numerical_grade", "predicted_grade"],
            var_name="type",
            value_name="Grade"
        )


        # Create a stacked bar plot
        plt.figure(figsize=(12, 6))
        bar_plot = sns.barplot(
            data=comparison_df,
            x="course_id",
            y="Grade",
            hue="type",
            palette="Set2"
        )


        # Add labels and title
        handles, labels = bar_plot.get_legend_handles_labels()

        plt.title("Comparison Between Real and Predicted Grade Percentages", fontsize=16)
        plt.xlabel("Courses", fontsize=12)
        plt.ylabel("Grade", fontsize=12)
        plt.legend(handles=handles, labels=["Real Grade", "Predicted Grade"], loc='upper right')
        plt.show()

      else:

        # Create a Seaborn bar plot
        plt.figure(figsize=(10, 6))
        sns.barplot(data=period_grades, x="course_id", y="numerical_grade", hue="course_id", palette="viridis")

        # Add labels and title
        plt.title(f"Grades for period {period}", fontsize=16)
        plt.xlabel("Courses", fontsize=12)
        plt.ylabel("Grade", fontsize=12)
        plt.show()



    return period_grades

  def get_grades_by_years(self, df: pd.DataFrame | None = None) -> pd.DataFrame:
    '''
    Get the DataFrame with the info of the grade performance over the years.

    This return a DataFrame you can either pass it as an argument for
    another method or simply convert it to JSON to sent as a API.
    '''
    if df is None:
      df = self.df

    avg_grades = df.groupby(["year", "period"]).agg({"numerical_grade": "mean", "predicted_grade": "mean"}).reset_index()

    if self.plotting:

      if self.use_predicted_values:

        # Preparing data for the double bar plot
        avg_grades["year_period"] = avg_grades["year"] + avg_grades["period"]


        comparison_df = pd.melt(
            avg_grades,
            id_vars=["year_period"],
            value_vars=["numerical_grade", "predicted_grade"],
            var_name="type",
            value_name="Grade"
        )


        # Create a stacked bar plot
        plt.figure(figsize=(12, 6))
        bar_plot = sns.barplot(
            data=comparison_df,
            x="year_period",
            y="Grade",
            hue="type",
            palette="Set2"
        )


        # Add labels and title
        handles, labels = bar_plot.get_legend_handles_labels()

        plt.title("Comparison Between Real Period Performance and Predicted performance", fontsize=16)
        plt.xlabel("Periods", fontsize=12)
        plt.ylabel("Grade", fontsize=12)
        plt.legend(handles=handles, labels=["Real Grade", "Predicted Grade"], loc='upper right')
        plt.show()

      else:

        # Create a Seaborn bar plot
        plt.figure(figsize=(10, 6))
        sns.barplot(data=avg_grades, x="year", y="numerical_grade", hue="period", palette="viridis")

        # Add labels and title
        plt.title("Average Grades Per Period Per Year", fontsize=16)
        plt.xlabel("Year", fontsize=12)
        plt.ylabel("Average Grade", fontsize=12)
        plt.legend(title="Period")
        plt.show()

    return avg_grades

````
